```python
challenge_1 = pd.read_sql_query("""
SELECT a.au_id AUTHOR_ID, a.au_lname LAST_NAME, a.au_fname FIRST_NAME, t.title TITLE, p.pub_name PUBLISHER 
FROM authors a 
JOIN titleauthor ta on ta.au_id = a.au_id
JOIN titles t on t.title_id = ta.title_id
JOIN publishers p on p.pub_id = t.pub_id
;
"""
, engine)
```

```python
challenge_2 = pd.read_sql_query("""
SELECT a.au_id AUTHOR_ID, a.au_lname LAST_NAME, a.au_fname FIRST_NAME, p.pub_name PUBLISHER, count(t.title) as TITLE_COUNT
FROM authors a
JOIN titleauthor ta on ta.au_id = a.au_id
JOIN titles t on t.title_id = ta.title_id
JOIN publishers p on p.pub_id = t.pub_id
GROUP BY AUTHOR_ID, p.pub_name
;

"""
, engine)
```

```python
challenge_2
```

```python
challenge_3 = pd.read_sql_query("""
SELECT a.au_id AUTHOR_ID, a.au_lname LAST_NAME, a.au_fname FIRST_NAME, count(t.title) as TOTAL
FROM authors a
JOIN titleauthor ta on ta.au_id = a.au_id
JOIN titles t on t.title_id = ta.title_id
GROUP BY AUTHOR_ID
ORDER BY TOTAL DESC
LIMIT 3
;
"""
, engine)
```

```python
challenge_3
```

```python
challenge_4 = pd.read_sql_query("""
SELECT a.au_id AUTHOR_ID, a.au_lname LAST_NAME, a.au_fname FIRST_NAME, count(t.title) as TOTAL
FROM authors a
LEFT JOIN titleauthor ta on ta.au_id = a.au_id
LEFT JOIN titles t on t.title_id = ta.title_id
GROUP BY AUTHOR_ID
ORDER BY TOTAL DESC
;
"""
, engine)
```
